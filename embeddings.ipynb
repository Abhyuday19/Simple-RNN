```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## Sentences
sent =[ 'the glass of milk',
         'the glass of juice',
         'the cup of tea',
         'understand the meaning of words',
         'you have done your work very well', ]
```

```python
sent
```

```python
## Defining the Vocabulary size
voc_size = 10000
```

```python
## One Hot Representation
## These below are the indexes which tell that which word of the first sentence is represented by which index Each word has it's own index representation.
one_hot_repr = [one_hot (words , voc_size) for words in sent]
one_hot_repr
```

```python
## Word Embedding Representation
## pad_sequence is used to make the length of the one hot representations of the same length.
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## if using the pre option , then all the zeros will come in the first during the padding.
## and if the post option is used then all the zeros will come at the end of the representations.
sent_length = 10
embeded_docs = pad_sequences(one_hot_repr ,padding='pre',maxlen=sent_length)
print(embeded_docs)
```

```python
## feature representations
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embeded_docs)
```
